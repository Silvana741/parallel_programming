# **IMPLEMENTATION 1 - CASE 1**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_records = [2000, 20000, 200000]
sequential_times = [13.169708967208862, 141.00727796554565, 1375.2240283489227]
# p = 2
parallel_times_2 = [8.851280689239502, 80.43314504623413, 749.9267830848694]
# p = 3
parallel_times_3 = [8.503436088562012, 77.37524819374084, 720.4661800861359]
# p = 4
parallel_times_4 = [7.070731163024902, 68.75394344329834, 696.1281995773315]
# p = 5
parallel_times_5 = [7.389263868331909, 70.98808598518372, 714.3846476078033]
# p = 6
parallel_times_6 = [7.3613011837005615, 72.32536029815674, 729.1205651760101]
# p = 7
parallel_times_7 = [7.3646605014801025, 71.57720303535461, 726.2418432235718]
# p = 8
parallel_times_8 = [7.64017915725708, 71.7573401927948, 800.9647336006165]
```

```python
result_data = np.array([sequential_times, parallel_times_2, parallel_times_3, parallel_times_4, 
                        parallel_times_5, parallel_times_6, parallel_times_7, parallel_times_8])
```

```python
result_data
```

```python
result_df = pd.DataFrame(result_data, columns=data_records)
```

```python
result_df
```

```python
numberOfProcesses = [i+1 for i in range(result_df.shape[0])]
numberOfProcesses
```

# **K = 5 with 2k records:**

---

```python
time_2k = result_df[2000]
time_2k
```

```python
table_data_2k = []
speed_2k = []
efficiency_2k = []
for p in numberOfProcesses:
  s = time_2k[0]/time_2k[p-1]
  e = s/p
  speed_2k.append(round(s,5))
  efficiency_2k.append(round(e,5))
  table_data_2k.append([p, round(s,5), round(e,5)])

print(speed_2k)
print(efficiency_2k)
print(table_data_2k)
```

```python
fig, axs = plt.subplots(2,1, figsize=(15,8), facecolor=(1, 1, 1))
# graph
graph = axs[0]
graph.grid()
graph.plot(numberOfProcesses, time_2k, marker='o')
graph.set_xlabel('No. processes')
graph.set_ylabel('Execution time (in seconds)')
graph.set_title('kNN = 5 with 2000 records')
for i in numberOfProcesses:
  x = i
  y = round(time_2k[i-1], 7)
  graph.annotate(y, xy=(x,y), 
               xytext=(20, 20), textcoords='offset points', 
               arrowprops=dict(arrowstyle="-", connectionstyle="angle,angleA=0,angleB=90,rad=5"))
# table
table = axs[1]
collabel=('No. processes', "Speed (S)", "Efficiency (E)")
table.table(cellText=table_data_2k,colLabels=collabel,loc='center')
table.axis('tight')
table.axis('off')
# show
plt.legend()
plt.show()
```
