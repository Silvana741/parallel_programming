# **IMPLEMENTATION 2 - CASE 1**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_records = [2000, 20000, 200000]
sequential_times = [15.511528253555298, 169.01578164100647, 2077.07509970665]
# p = 2
parallel_times_2 = [8.95763897895813, 89.94237408638, 1156.5261626243591]
# p = 3
parallel_times_3 = [8.854078578948975, 89.31939649581909, 1156.2500643730164]
# p = 4
parallel_times_4 = [8.534229040145874, 89.0028510093689, 1006.4967381954193]
# p = 5
parallel_times_5 = [8.569545030593872, 83.92872381210327, 869.1917464733124]
# p = 6
parallel_times_6 = [8.555257081985474, 83.26732587814331, 865.2941389083862]
# p = 7
parallel_times_7 = [8.880664587020874, 82.99138498306274, 870.9944756031036]
# p = 8
parallel_times_8 = [8.890645217895508, 83.99714541435242, 847.7153644561768]
```

```python
result_data = np.array([sequential_times, parallel_times_2, parallel_times_3, parallel_times_4, 
                        parallel_times_5, parallel_times_6, parallel_times_7, parallel_times_8])
```

```python
result_data
```

```python
result_df = pd.DataFrame(result_data, columns=data_records)
```

```python
result_df
```

```python
numberOfProcesses = [i+1 for i in range(result_df.shape[0])]
numberOfProcesses
```

# **K = 5 with 2k records:**

---

```python
time_2k = result_df[2000]
time_2k
```

```python
table_data_2k = []
speed_2k = []
efficiency_2k = []
for p in numberOfProcesses:
  s = time_2k[0]/time_2k[p-1]
  e = s/p
  speed_2k.append(round(s,5))
  efficiency_2k.append(round(e,5))
  table_data_2k.append([p, round(s,5), round(e,5)])

print(speed_2k)
print(efficiency_2k)
print(table_data_2k)
```

```python
fig, axs = plt.subplots(2,1, figsize=(15,8), facecolor=(1, 1, 1))
# graph
graph = axs[0]
graph.grid()
graph.plot(numberOfProcesses, time_2k, marker='o')
graph.set_xlabel('No. processes')
graph.set_ylabel('Execution time (in seconds)')
graph.set_title('kNN = 5 with 2000 records')
for i in numberOfProcesses:
  x = i
  y = round(time_2k[i-1], 7)
  graph.annotate(y, xy=(x,y), 
               xytext=(20, 20), textcoords='offset points', 
               arrowprops=dict(arrowstyle="-", connectionstyle="angle,angleA=0,angleB=90,rad=5"))
# table
table = axs[1]
collabel=('No. processes', "Speed (S)", "Efficiency (E)")
table.table(cellText=table_data_2k,colLabels=collabel,loc='center')
table.axis('tight')
table.axis('off')
# show
plt.legend()
plt.show()
```
